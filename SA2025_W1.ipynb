### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
df_reset = df.reset_index()
grouped = df_reset.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped, x='model_year', y='mpg', hue='origin')
plt.title("Average MPG Over Years by Origin")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### We define a â€œconsistentâ€ car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Reset index to get 'name' column back
df_with_name = df_reset.copy()
if 'name' not in df_with_name.columns:
  # If 'name' is not present, read from original CSV
  df_with_name = pd.read_csv("Cars.csv")

# Group by 'name' and count unique model_years
name_year_counts = df_with_name.groupby('name')['model_year'].nunique()

# Filter to names with more than one model_year
multi_year_names = name_year_counts[name_year_counts > 1].index

# For each such name, compute std(mpg), count, and mean(mpg)
stats = (
  df_with_name[df_with_name['name'].isin(multi_year_names)]
  .groupby('name')
  .agg(
    mpg_std=('mpg', 'std'),
    appearances=('mpg', 'count'),
    avg_mpg=('mpg', 'mean')
  )
)

# Filter for std(mpg) < 1.0
consistent = stats[stats['mpg_std'] < 1.0]

# Find the minimum std(mpg)
min_std = consistent['mpg_std'].min()

# Get car(s) with the lowest variation
result = consistent[consistent['mpg_std'] == min_std]

# Bonus: sort by appearances (desc), then avg_mpg (desc)
result = result.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

# Show model name(s), number of appearances, and average mpg
result[['appearances', 'avg_mpg']]
```
